```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

arr = pd.read_csv('relusts.csv')
size = 14 #lenght of 1 data sample
data = [arr.loc[i:i+size-1,:] for i in range(0, len(arr),size)]
print(data[1]['selector_method'].iloc[0])

for funct in data:
    label = funct['selector_method'].iloc[0] + " "+ funct['classifier'].iloc[0] + " " + funct['balance_method'].iloc[0]
    plt.scatter(funct['k'],funct['accuracy'], label=label, linewidths=0.01)
    plt.plot(funct['k'],funct['accuracy'],linewidth=0.5)
plt.legend(shadow=True, fancybox=True,bbox_to_anchor=(1.04, 1), loc="upper left")
```
